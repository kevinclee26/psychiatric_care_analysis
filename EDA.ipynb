```python
import pandas as pd
```

```python
df=pd.read_csv('data/Inpatient_Psychiatric_Facility_Quality_Measure_Data___by_Facility.csv')
df.head()
```

```python
columns=df.columns
keep_cols=[]
for each_col in columns: 
    if not 'Description' in each_col: 
        keep_cols.append(each_col)
clean_df=df.loc[:, keep_cols] # df.loc[:, [col1, col2, col3]]
```

```python
clean_df.dropna()
```

```python
clean_df.dtypes
```

```python
for idx, row in clean_df.iterrows(): 
    for sidx, cell in row.iteritems(): 
        clean_df.loc[idx, sidx]=str(cell).replace('%', '')
```

```python
clean_df.to_csv('data/clean_data.csv', index=False)
```
